## BONUS:

Below is a python code to draw the famous fractal which you can read by clciking on: [Mandelbrot Set](https://en.wikipedia.org/wiki/Mandelbrot_set).

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import newaxis

#to ignore warnings
import warnings
warnings.filterwarnings("ignore")

def compute_mandelbrot(N_max, threshold, nx, ny):
    
    # A grid of c-values
    x = np.linspace(-2, 1, nx)
    y = np.linspace(-1.5, 1.5, ny)
    c = x[:,newaxis] + 1j*y[newaxis,:]
    
    # Mandelbrot iteration
    z = c
    for j in range(N_max):
        z = z**2 + c
        mandelbrot_set = (abs(z) < threshold)
    
    return mandelbrot_set
```

- You may create different visualization stages stages by controling static number of iterations per pixel by controlling the $x_i$ values in 
> compute_mandelbrot($x_1$, $x_2$, $x_3$, $x_4$)

```python
mandelbrotSet = compute_mandelbrot(50, 50., 600, 400)
plt.imshow(mandelbrotSet.T, extent=[-2, 1, -1.5, 1.5])
plt.show()
```
