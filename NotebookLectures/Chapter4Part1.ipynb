## Three-dimensional plots 
- A two dimensional object: when you cut through a two dimensional object, for example a curved paper, you obtain a curve or a line. 
- We need three axises to plot dataset in a space $\mathbb{R}^3$. Thus, let's introduce the 3D space with `ax = Axes3D(fig)` available from the `mpl_toolkits.mplot3d` module.
- We can also use `add_subplot` function with argument `projection ="3d"`

### Spatial Curve

 - Let's create a Helix space curve, $H(t) = (x(t), y(t), z(t))$, which looks like a spring. This can can be expressed in the form a parametric curve as: 
 $$
 x(t) = r \cos(t) \\
 y(t) = r \sin(t) \\
 z(t) = r t
 $$
 where $t \in [0, 2\pi]$ is a cycle, with radius and constant $r \in \mathbb{R}$ respectively.

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2, figsize=(14, 4), subplot_kw={'projection': '3d'})

# Plot a helix curve in the form of parametric

r = 2 #radius

t = np.linspace(0, 6* np.pi, 100)

x = r * np.cos(t)
y = r * np.sin(t)
z = r * t

ax[0].scatter(x, y, z, zdir='z', label='Helix')
ax[0].set_title("Scatter Plot")

ax[1].plot(x, y, z, zdir='z', label='Helix')
ax[1].set_title("Standard Plot")

# Make legend, label axes.

for x in ax:
    x.legend()
    x.set_xlabel('X')
    x.set_ylabel('Y')
    x.set_zlabel('Z')
    x.view_init(elev=20., azim=-35)# Customize the view angle

plt.show()
```

### Surface Plot

Let a surface which looks like waves in space be defined in a explicit form as follows; 
$$
S(x,y) = \frac{a \cos (a x) \sin (b y)}{b}
$$  
where $x,y \in [0, 2\pi]$, with constants $a, b \in \mathbb{R}$ respectively.

```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

x = y = np.linspace(0, 2*np.pi, 74)
X, Y = np.meshgrid(x, y)

a = -0.9
b = 0.6

Z = (a * np.cos(a * X) * np.sin(b * Y)) / b
```

- We draw 3D graphs with `Axes3D` object which is available in `mpl_toolkits.mplot3d` module.

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure()
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)
ax.plot_surface(X, Y, Z, rstride=10, cstride=10, cmap=cm.viridis)
ax.view_init(elev=25., azim=-65)
```

- We can also directly draw 3D plot by using `plt.subplot()` with argument `subplot_kw={'projection': '3d'}`

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4), subplot_kw={'projection': '3d'})

def title_and_labels(ax, title):
    ax.set_title(title)
    ax.set_xlabel("$x$", fontsize=16)
    ax.set_ylabel("$y$", fontsize=16)
    ax.set_zlabel("$z$", fontsize=16)

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())

p = axes[0].plot_surface(X, Y, Z, rstride=5, cstride=5, linewidth=1, antialiased=False, norm=norm, cmap=mpl.cm.Blues)
#cb = fig.colorbar(p, ax=axes[0], shrink=0.9)
title_and_labels(axes[0], "Surface Plot")

axes[1].plot_wireframe(X, Y, Z, rstride=5, cstride=5, color="grey")
title_and_labels(axes[1], "Wireframe Plot")

cset = axes[2].contour(X, Y, Z, zdir='z', offset=-1.5, norm=norm, cmap=mpl.cm.Greens)
cset = axes[2].contour(X, Y, Z, zdir='y', offset=6, norm=norm, cmap=mpl.cm.Reds)
title_and_labels(axes[2], "Contour")

fig.tight_layout()
fig.savefig("3d-plots.png", dpi=200)
```
