## Time series 

- with Pandas, Seaborn & plotnine

- Read more [Working with pandas: Time series](https://pandas.pydata.org/docs/getting_started/intro_tutorials/09_timeseries.html)

-  In mathematics, a [time series](https://en.wikipedia.org/wiki/Time_series) is a series of data points indexed (or listed or graphed) in time order. Most commonly, a time series is a sequence taken at successive equally spaced points in time. Thus it is a sequence of discrete-time data. Examples of time series are daily rainfall and the daily closing value of the KLCI index.

- In a broader context, a signal which may change over time (temporal signal) or as a function of space coordinates (spatial signal). Rainfalls can be thought as temporal signal in one-dimension and black & white images as spatial signal in two dimensions.

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3 , figsize=(8, 4))

sns.boxplot(data = populations2, ax=ax1, palette="pastel")
sns.violinplot(data = populations2, ax=ax2, palette="vlag")
sns.stripplot( data=populations2, ax=ax1, size=4, color=".3", linewidth=0)
sns.stripplot( data=populations2, ax=ax3, size=4, color=".3", linewidth=0)

fig.tight_layout()
#fig.savefig("seaborn-boxplot-violinplot.pdf")
```

- Visualizing  relationships between two variables

```python
sns.relplot(x="hare", y="carrot", data=populations2);
```

```python
sns.relplot(x="hare", y="lynx", data=populations2);
```

```python
sns.relplot(x="lynx", y="carrot",kind="line", 
            height=3, aspect=1, linewidth=1.5,  data=populations2);
```

```python
sns.relplot(x="hare", y="lynx", hue="carrot",
            data=populations2)
```

```python
sns.relplot(x="hare", y="lynx", size="carrot",
            data=populations2)
```

- Assigning a second variable to y, however, will plot a bivariate distribution.
- A bivariate histogram bins the data within rectangles that tile the plot and then shows the count of observations within each rectangle with the fill color (analagous to a heatmap()).

```python
sns.displot(populations2, x="hare", y="lynx", cbar=True)
```

-Similarly, a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian. The default representation then shows the contours of the 2D density:

```python
sns.displot(populations2, x="hare", y="lynx", kind="kde", rug=True)
```

```python
g = sns.PairGrid(populations2)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
```

## Using `plotnine`

---
`ggplot` is a Python implementation of the book **Grammar of Graphics** which is implemented in R (ggp2plot).
- **The basic building blocks are of [Grammar of Graphics](https://link.springer.com/book/10.1007%2F0-387-28695-0) are:** 

1. **data:** The data + a set of aesthetic mappings that describing variables mapping
2. **geom:** Geometric objects, represent what you actually see on the plot: points, lines, polygons, etc.
3. **stats:** Statistical transformations, summarise data in many useful ways.
4. **scale:** The scales map values in the data space to values in an aesthetic space
5. **coord:** A coordinate system, describes how data coordinates are mapped to the plane of the graphic.
6. **facet:** A faceting specification describes how to break up the data into subsets for plotting individual set


`plotnine` is a Grammar of Graphics implementation for Python by Hassan Kibirige.
